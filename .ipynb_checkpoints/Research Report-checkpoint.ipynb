### <span style='font-family:"Times New Roman"'> <span styel=''>Background   
#### <span style='font-family:"Time New Roman"'> <span style=''>Tumor Heterogeneity   
<span style='font-family:"Time New Roman"'> <span style=''> Tumor heterogeneity is a state in which cancer cells within the same tumor are fundamentally different from each other or in different cell states. Sequencing technologies have confirmed this hypothesis by discovering subclonal cell populations within tumors. As a patient’s cancer progresses, these cell populations–genetically distinct groups of tumor cells descending from a common ancestor–can be identified and their evolutionary trajectory can be followed with response and resistance to treatment. 

<span style='font-family:"Time New Roman"'> <span style=''> This hypothesis was first made in 1976 with Peter Nowell who suggested that tumors have multiple heterogenous sub clones. Because of this, individualized cancer treatment emerged due to the desire to stop treating cancer as if it was homogenous. Specifically, intra-tumor heterogeneity is closely related to cancer progression, resistance to therapy, and cancer recurrences. Inter-tumor heterogeneity exceeds boundaries of specific tumors and their molecular classifications which makes clinical treatment a challenge. 
    
    
<span style='font-family:"Time New Roman"'> <span style=''> Tumor heterogeneity consists of two types: intertumor (tumor by tumor) and intratumor (within a tumor) heterogeneity. Intertumor heterogeneity refers to PLCs from different patients whose altered genotype/phenotype are caused by external factors (etiological and environmental factors). Opposingly, intratumor heterogeneity refers to genomic/biological variations within a tumor gained by tumor cell evolution under internal factors (microenvironments linked to different etilogies). A tumor lesion typically evolves from a single malignant cell into a functionally heterogenous tumor mass with a hierarchically organized tumor cell community, promoting its survival in response to various microenvironments. 

#### <span style='font-family:"Times New Roman"'> <span styel=''>Single Cell RNA Sequencing 
<span style='font-family:"Times New Roman"'> <span styel=''>Single cell RNA sequencing (scRNAseq) has dominated bioinformatics research with the invaluable information it provides such as complex and rare cell populations, regulatory relationships between genes, and the trajectories of distinct cell lineages in development. Cell information is stored within the DNA which is transformed into RNA and then into proteins. Proteins fuel everything in the cell. However, biologists are not able to quantify the exact amount of protein in each cell, but sequencing the RNA is possible. The reasoning for quantifying RNA is that different cells may express different genes on different levels. These differences are a key into the behavior of cells. The absolute quantity of these molecules is very small in a single cell, so experiments often consist of billions of cells. When RNA is extracted from a group of cells, all information about the difference between them is lost, so the data is actually a mean expression of each gene. Because of this, whenever an RNA molecule is sequenced, a unique technical sequence is also sequenced. Additionally, this helps identify the real number of RNA in each cell. For each sequenced molecule, the information attained is the ID of the cell–called the barcode–and it's own unique ID–called the unique molecular identifier (UMI). The resulting data from the sequencing is formatted in a genes-by-cells table P, where P[i, j] is an expression of genes i in cell j. 

<span style='font-family:"Times New Roman"'> <span styel=''>The major use of scRNAseq has been to assess transcriptional (protein-encoding, mRNA molecules) similarities and differences within a population of cells with reports revealing undetected levels of heterogeneity in cell populations. For example, scRNAseq can be used to identify malignant tumour cells within a tumour mass or hyper-responsive immune cells in a seemingly homogenous group of cells. The success of heterogeneity analysis has prompted many researchers to embark on scRNaseq studies. This can also provide important information about gene expression. Studying these gene expression patterns at the single-cell level can lead to observation of gene-regulatory networks that underlie heterogeneity and cell-type specification. 
    
<span style='font-family:"Times New Roman"'> <span styel=''>The common workflow is (1) isolation of single cells (2) cell lysis while preserving mRNA (3) mRNA capture (4) reverse transcription of primed RNA into ocmplementary DNA (5) cDNA amplification (6) preparation of cDNA sequencing library (7) pooling of sequence libraries. After those steps, the sequencing data is analyzed through bioinformatics and computational means. The end result of scRNAseq data is that it can be clustered and then differentially expressed genes (most expressed genes) are identified within the clusters providing biomarkers for targeted therapy development.

#### <span style='font-family:"Times New Roman"'> <span styel=''> Machine Learning   
<span style='font-family:"Times New Roman"'> <span styel=''> Machine learning is a modern computational tool that is the cross section of advanced mathematical topics such as linear algebra and computer science. Machines–or in this case computers–automatically learn and improve from experience without being explicitly programmed. In the large umbrella of machine learning, there are many sects which stem from the two types: supervised and unsupervised learning. Supervised learning is primarily conducted when the data has labels or more specifically when there is a target variable on a given set of predictors. Unsupervised learning is primarily conducted when the data lacks labels and/or a target variable on a given set of predictors. Given the unlabeled nature of scRNAseq data as well as the lack of independent variables, this is a textbook problem for unsupervised learning analysis. 
    
<span style='font-family:"Times New Roman"'> <span styel=''> Unsupervised learning is done with cluster analysis. Clustering is the process of dividing the entire data into groups–the clusters–based on patterns within the data (in this case gene expression). In optimal clusters, all data points should be close to each other and the groups should be separated to indicate highly diverse groups. With scRNAseq data, biologists often question given K cells, how many different cell states are there? The cell states are represented through the clusters.

<span style= 'font-family: "Time New Roman"'> **6. Dimensionality Reduction with PCA** 

<span style= 'font-family: "Time New Roman"'> A Scree Plot is a diagnostic tool to check whether PCA works well on the current data. Principal components are made in the order of the amount of variation they cover: the first component captures the most and it decreases from there. In PCA, there are as many principle components as features. The scree plot should be an indicator of what principal components to keep; ideally, the graph should somewhat look like an elbow, and the start of the elbow is the cut off (or the maximum number of principal components to include). The y axis is eigenvalues which means the amount of variation kept with each component. The kept principal components should be able to represent at least 80% of the total variance in the dataset. 

```python
U, S, V = np.linalg.svd(mat) 
eigvals = S**2 / np.sum(S**2)  
fig = plt.figure(figsize=(10, 6))
sing_vals = np.arange(len(mat)) + 1
plt.plot(sing_vals, eigvals, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
```

<span style= 'font-family: "Time New Roman"'> The total number of features in the dataset is 600, so naturally our scree plot is very zoomed out. However, minor skews can be observed around 30-50 components. For a better view or somewhat zoomed in scree plot, a sub section of the data may be more useful. 


<span style= 'font-family: "Time New Roman"'> More skews can be seen with a light elbow at 50 components. That will be the parameter to use.
    
<span style= 'font-family: "Time New Roman"'>PCA is a dimensionality reduction technique which performs feature extraction on a dataset. Feature extraction creates "new" independent variables from selected independent variables that combine each of the old variables and order them by how well they predict the dependent variable. From there, dimensions are reduced based on which important new independent variables are kept. In the end, the most valuable parts of all previous variables are kept when the most important new independent variables are selected. 

<span style= 'font-family: "Time New Roman"'>PCA works by calculating a matrix that summarizes how each of the variables relate to one another. This matrix breaks into two components: direction and magnitude. Then, the data is aligned with the important directions. With direction selection then data compression, the dimensionality of the feature space is reduced; however, the data is only transformed into different directions, so all original variables are still in the model. 

```python
pca = PCA(n_components=50)
pca.fit(mat)
mat_reduce=pca.transform(mat)
```

<span style= 'font-family: "Time New Roman"'> **7. Preliminary Visualization with UMAP** 

<span style= 'font-family: "Time New Roman"'> Now that the data has gone through feature extraction and the most variable components are kept, the data can be projected with UMAP. UMAP stands for Uniform Manifold Approximation and Projection for Dimension Reduction. It provides very general framework for approaching manifold learning and dimension reduction. It is mainly used for topological data analysis. 

<span style= 'font-family: "Time New Roman"'> The way the UMAP algorithm works is that it takes all the data points {X <sub>1</sub>,....,X<sub>N</sub>} and then its *k* nearest neighbors. Let *L<sub>i</sub>* be the diametrs of the neighborhood of X<sub>i</sub> and let p<sub>i</sub> be the distance from X<sub>i</sub> to its nearest neighbor. Then, a weighted graph is formed by taking X<sub>j</sub> in the k-nearest neighborhood of X<sub>i</sub> and defining w<sub>i</sub>(X<sub>i</sub>, X<sub>j</sub>) = exp(-(*d*(X<sub>i</sub>, X<sub>j</sub>)-p<sub>i</sub>)/*L<sub>i</sub>*. To symmetrize it, w(X<sub>i</sub>, X<sub>j</sub>)= w<sub>i</sub>(X<sub>i</sub>, X<sub>j</sub>) + w<sub>j</sub>(X<sub>i</sub>, X<sub>j</sub>) - w<sub>i</sub>(X<sub>i</sub>, X<sub>j</sub>)w<sub>j</sub>(X<sub>i</sub>, X<sub>j</sub>) is used. Then, given two weights w and w' on the dataset, cross entropy is represented by: 

w represents the weights computed from the dataset and w' represents the weights computed from our low-dimensional embedding. The cross entropy is the sum of uncoupled terms, one for each edge, so stochastic gradient descent is applied by choosing one term to approximate the gradient at each step. The weights w'(i, j) depend only on the points withint he neighborhoods of X<sub>i</sub> and X<sub>j</sub> which is 2k-1 points. That is the entirety of the algorithm. 

<span style= 'font-family: "Time New Roman"'> For the purposes of summarization, UMAP simply constructs a high dimensional graph representation of the data then optimizes a low-dimensional graph to be as structurally similar as possible. The underlying mathematical concepts described above are how UMAP makes its high dimensional graph. Then, UMAP makes the graph "fuzzy" by decreasing the likelihood of connection as the radius grows. By stipulation that each point must be connected to at least its closest neighbor, UMAP preserves local structure in balance with global structure. 

```python
embedding = umap.UMAP(n_neighbors=30, min_dist=0.01, metric='euclidean').fit_transform(mat_reduce)
plt.figure(figsize=(5, 5))
plt.scatter(embedding[:,0],embedding[:,1], s=50.0, color='red')
plt.title('Naive Clustering')
```

<span style= 'font-family: "Time New Roman"'> **8. Testing of Multiple Clustering Algorithms for Consistency, Evaluation, and Analysis Purposes** 

<span style= 'font-family: "Time New Roman"'> Now that principal components have been selected and dimensions have been reduced, multiple clustering algorithms can be tested for not only data analysis but also for a consistency evaluation. If the clusters are consistent, that's an indicator that the parameters used are optimal. The parameters have already been fine tuned for PCA using the Scree Plot as well as UMAP with trial and error as well as observation. 8 clustering algorithms will be tested: Affinitiy Propagation, Mean Shift, Spectral Clustering, Ward, Agglomerative Clustering, DBSCAN, Birch, and the Gaussian Mixture Model. These are all commonly used clustering algorithms, and they provide a variety of graph based clustering, hierarchal clustering, and mixture models to choose from. 

<span style= 'font-family: "Time New Roman"'> The first step is to define all of the parameters into a dictionary such as quantile (where a sample is divided into adjacent subgroups), epochs (one complete presentation of the dataset to be learned to a learning machine), damping (measure to describe how rapidly oscillations decay from one bounce to the next), preference (task of learning to predict an order relation on collection), n_neighbors (how many points are assumed to be connected to one point), and n_clusters (the amount of clusters differentiated). 

```python
params = {'quantile': .3, 'eps': .3, 'damping': .9, 'preference':-200, 'n_neighbors':30, 'n_clusters':3}
bandwidth = estimate_bandwidth(embedding, quantile=params['quantile'])
connectivity = kneighbors_graph(embedding, n_neighbors=params['n_neighbors'], include_self = False)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ward = AgglomerativeClustering(n_clusters=params['n_clusters'], linkage='ward', connectivity=connectivity)
spectral = SpectralClustering(n_clusters=params['n_clusters'], eigen_solver='arpack', affinity="nearest_neighbors")
dbscan = DBSCAN(eps=params['eps'])
affinity_propagation = AffinityPropagation(damping=params['damping'], preference=params['preference'])
average_linkage = AgglomerativeClustering(linkage="average", affinity="cityblock", n_clusters=params['n_clusters'], connectivity=connectivity)
birch = Birch(n_clusters=params['n_clusters'])
gmm = GaussianMixture(n_components=params['n_clusters'], covariance_type='full')
```

<span style='font-family:"Time New Roman"'> Now it's time to plot the data!

```python
clustering_algorithms = (
    ('AffinityPropagation', affinity_propagation),
    ('MeanShift', ms),
    ('SpectralClustering', spectral),
    ('Ward', ward),
    ('AgglomerativeClustering', average_linkage),
    ('Birch', birch),
    ('DBSCAN', dbscan),
    ('GaussianMixture', gmm))
f, ax = plt.subplots(2, 4, figsize=(25,15))
for idx, (name, algorithm) in enumerate(clustering_algorithms):
    algorithm.fit(embedding)
    if hasattr(algorithm, 'labels_'):
        y_pred = algorithm.labels_.astype(np.int)
    else:
        y_pred = algorithm.predict(embedding)
    colors = np.array(list(islice(cycle(['black', 'red', 'green']),
                                    int(max(y_pred) + 1))))
    colors = np.append(colors, ["#000000"])
    ax[idx//4, idx%4].scatter(embedding[:, 0], embedding[:, 1], s=2, color=colors[y_pred])
    ax[idx//4, idx%4].set_xticks(())
    ax[idx//4, idx%4].set_yticks(())
    ax[idx//4, idx%4].set_title(name)
```

<span style= 'font-family: "Time New Roman"'> The most consistent and defined clusters are from Spectral Clustering  and DBSCAN. 
    
<span style= 'font-family: "Time New Roman"'> **Spectral Clustering**: a graph based clustering technique where the approach is used to identify communities of nodes in a graph plotted based on the edges connecting them. 
    
<span style= 'font-family: "Time New Roman"'> **DBSCAN**: (Density-Based Spatial Clustering of Applications with Noise) a spatial clustering method which groups data together based on the similarity of observations. 

<span style='font-family: "Time New Roman"'> **9. K-Means Clustering**

<span style='font-family: "Time New Roman"'> K-Means clustering another common method of clustering. The algorithm works by computing a centroid (center) of each cluster and then minimizing the sum of distances of all points within the cluster to the centroid. It does this by first selecting a random data point as the first center of the cluster and then re-computes the centroid until the difference between the previous centroid and the new centroid stops changing. 

<span style='font-family: "Time New Roman"'> This approach will create a 3D plane for clustering which more accurately displays the highly dimensional data. 

<span style= 'font-family: "Time New Roman"'> First, we will create a dataframe from the dataset, filter it, and then plot a dendrogram to see some preliminary hierarchal clustering. 

```python
datafile = 'dataset2.count.txt'
df = pd.read_csv(datafile, sep='\t')
df = df.set_index('Gene.Symbol')
cells = list(df.columns.values)
mat = np.array(df)
per_cell_sum = mat.sum(axis=1)
per_gene_sum = mat.sum(axis=0)
low_expr_thr = 100
high_expr_thr = 1000000
mat = mat[:, (per_gene_sum>=low_expr_thr) & (per_gene_sum<=high_expr_thr)]
mean_exp = mat.mean(axis=0)
std_exp = np.sqrt(mat.std(axis=0))
CV = std_exp/mean_exp
mat = mat[:, CV >= 1.5]
cells_expression = mat.sum(axis=1)
mat = mat[cells_expression>=100, :]
mat = np.log(mat+1)
def dendrogrammer(df, leaf_labels):
    D = df.values
    if len(leaf_labels) != len(D):
        D = np.transpose(D)
    Z = linkage(D, method='ward', metric='euclidean')
    plt.figure(figsize=(10, 10))
    ax = plt.subplot()
    plt.subplots_adjust(left=0.07, bottom = 0.3, right = 0.98, top = 0.95, wspace=0, hspace=0)
    plt.xlabel('Cell Line')
    plt.ylabel('Distance')
    
    dendrogram(Z, leaf_rotation=90., leaf_font_size=10., labels = leaf_labels)
    plt.savefig('Dendrogram RNA.png')
df = pd.DataFrame(mat,index=mat[:,0])
dendrogrammer(df, cells)
```

<span style= 'font-family: "Time New Roman"'> The dendrogram does not explain a lot, but gives a basic visual of what cells have the same global expression profiles. 

```python
def PCAer(df):
    D = df.values
    D = np.transpose(D)
    pca = PCA(n_components= 50)
    pca.fit(D)
    projected = pca.fit_transform(D)
    return projected
def kmeanser(df, k): 
    Dpc = PCAer(df)
    kmeans = KMeans(n_clusters=k)
    kmeans = kmeans.fit(Dpc)
    labels = kmeans.predict(Dpc)
    C = kmeans.cluster_centers_
    out = [Dpc, labels, C]
    return out
[proj, labels, centroids] = kmeanser(df, 3)
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1 = Axes3D(fig1)
ax1.scatter(proj[:,0], proj[:, 1], proj[:, 2], c=labels, s=50, cmap='Accent')
ax1.scatter(centroids[:,0], centroids[:, 1], centroids[:,2], c = 'red', s=100, marker='*', cmap='Accent')
```

<span style= 'font-family: "Time New Roman"'> For evaluation, the elbow method will be used. The elbow method is a line plot of the varying inertia calculated for different values of k (different numbers of clusters). The most optimal k is determined at the point at which the line plot has an elbow shape or bends. 

```python
from scipy.spatial.distance import cdist
distortions = []
K = range(1, 10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(df)
    kmeanModel.fit(df)
    distortions.append(kmeanModel.inertia_)
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

<span style= 'font-family: "Time New Roman"'> Both k=2 and k=3 look good. However, due to biological knowledge of gene expression patterns as well as cell state, it is known that k=3 is the amount of clusters needed; this will be explained later in the results section.

### <span style= 'font-family: "Time New Roman"'> Conclusion 

<span style= 'font-family: "Time New Roman"'> To review, the goal of this research was to examine different machine learning classifiers and their effectiveness on identifying sub clonal cell populations. With more knowledge of what sub clonal cell populations truly entails, it was discovered that this simply means groups of cells with different gene expression profiles that can be differentiated based on state. This analysis would then be interpreted by biologists to find marker genes to target with immunotherapy. The machine learning algorithm would be deemed successful if it could group cells with similar gene profiles into different clusters with accuracy. The accuracy would come from (1) making the correct subdivisions (2) visual consistency and lack of extreme outliers (3) approximate or equivalent totals to that of the truth data. 

<span style= 'font-family: "Time New Roman"'> For the first criterion, all of the ML clustering algorithms were able to make the correct number of subdivisions, for the most part. The only exception would be with some clustering algorithms which would have a few outliers, or algorithms that would make divisions within a cluster. Despite these rare cases, most performed well. Although this is mainly due to the n_clusters parameter when computing the cluster plots, the tuning of hyperparameters in UMAP and PCA also led to the correct amount of subdivisions in a population. With this foundation, it is shown that machine learning is able to accurately divide a total cell population into the correct amount of subclonal cell populations. 
    
<span style= 'font-family: "Time New Roman"'> For the second criterion, approximately half of the ML clustering algorithms were able to be visually consistent with one another and have well defined clusters based on color and data point separation. The K-Means clustering algorithm was able to do this as well, for it scored well on the elbow plot for the correct value of k which signaled its clusters had good inertia. The poorly performing ML clustering algorithms for this criterion had the same general topological structure but lacked organization within the local clusters, having multiple colored groups and some outlier data points. 
    
<span style= 'font-family: "Time New Roman"'> Lastly for the third criterion, there were three outstanding algorithms which accurately classified cell state and molecular subtypes within the population based on numerical comparison to the truth data. For the RNAMAT dataset, K-Means clustering had almost equivalent results give or take a few cells to the truth data. K-Means successful performance can be explained due to its reliable methods of basing clusters off of their center as well as its ability to cluster data in three dimensions; with more dimensions to the data, less features were lost in the dataset allowing for more classification for each cell. For DATASET4, two algorithms worked extremely well: Ward and Birch clustering. Both had almost identical subclonal cell populations compared to the truth data. Ward and Birch clustering algorithms are both hierarchal clustering algorithms meaning that they both group clusters based on high similarity in an attempt to build a hierarchy of clusters. Due to the vastly different gene expression profiles of cells in DATASET4 which took gene counts per million, hierarchal clustering would have a higher chance of a more successful performance. 

<span style= 'font-family: "Time New Roman"'> In conclusion, machine learing is deemed successful in identifying subclonal cell populations within a vast cell population on the basis of accurate subdivisions, visual consistency, and approximate subclonal cell population size. The K-Means clustering algorithm performed incredibly well on very high dimensional data, and hierarchal clustering (Birch and Ward) performed extremely well on scRNAseq data with vastly diverse gene expression profiles. However, even preliminary UMAP visualization with parameter tuning was able to represent topological structure of cell populations before advanced clustering algorithms. Many clustering algorithms also yielded visually consistent results proving that ML is extremely helpful for cluster analysis. Overall, this research showed the potential accuracy K-Means, Birch, Ward, and other unsupervised learning algorithms can achieve in combination with UMAP and PCA on scRNAseq data analysis for identification of subclonal cell populations, which can later be used for immunotherapy development against differentially expressed genes in each of the subpopulations. 

### <span style= 'font-family: "Time New Roman"'> Further Research 


<span style= 'font-family: "Time New Roman"'> Machine learning has been proven to be effective in scRNAseq data analysis, for it can clearly distinguish between sub clonal cell populations with very similar population sizes compared to the truth data. The next step to processing the data to put it to use towards targeted therapy is finding the genes within in each cluster. For this task, the differentially expressed genes need to be identified, or rather biomarkers for each cluster need to be identified. For this, one would need to research literature for known marker genes and receive input from biologists. Then, there is the Seurat library in R or the ScanPy library in Python made for finding differentially expressed genes in datasets. These work by comparing the gene expression profiles within each cluster. 

<span style= 'font-family: "Time New Roman"'> Targeted therapy is a cancer treatment that targets specific genes and proteins involved in the growth of cancer within a patient. With the discoverey of heterogeneity within the cancer cell population and gene identifiers, targeted therapy can be developed to target the differentially expressed genes that are growing resistant to treatment.
